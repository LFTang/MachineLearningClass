```python
target = train['SalePrice']
train.drop('SalePrice', axis=1, inplace=True)
```

```python
train = pd.get_dummies(train)
```

```python
train = pd.get_dummies(train, columns = {'MSSubClass'})
```

```python
X_train, X_valid, y_train, y_valid = train_test_split(train, target, test_size=0.2, random_state=123)
```

```python
X_train
```

```python
X_train = X_train.fillna(X_train.mean())
X_valid = X_valid.fillna(X_valid.mean())
train = train.fillna(train.mean())
```

```python
from sklearn.preprocessing import scale
from sklearn.linear_model import Ridge, RidgeCV, Lasso, LassoCV
from sklearn.metrics import mean_squared_error

clf = Ridge(alpha=10)
clf.fit(X_train, y_train)
```

```python
alphas = 10**np.linspace(10,-2,100)*0.5
ridge = Ridge(normalize=True)
coefs = []
for a in alphas:
    ridge.set_params(alpha=a)
    ridge.fit(X_train, y_train)
    coefs.append(ridge.coef_)
np.shape(coefs)
```

```python
import matplotlib.pyplot as plt
ax = plt.gca()
ax.plot(alphas, coefs)
ax.set_xscale('log')
plt.axis('tight')
plt.xlabel('alpha')
plt.ylabel('weights')
plt.show()
```

```python
ridge2 = Ridge(alpha=.1, normalize=False)
ridge2.fit(X_train, y_train) # Fit a ridge regression on the training data
pred2 = ridge2.predict(X_valid) # Use this model to predict the test data
print(pd.Series(ridge2.coef_, index=X_train.columns)) # Print coefficients
print(np.sqrt(mean_squared_error(y_valid, pred2)))
```
